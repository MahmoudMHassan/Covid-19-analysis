```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print('Modules are imported.')
```

Importing the covid19 dataset 

```python
corona_dataset_csv = pd.read_csv("covid19_Confirmed_dataset.csv")
corona_dataset_csv.head() #returns first 5 rows of the dataframe
```

```python
corona_dataset_csv.shape #returns a tuple of rows and columns of a dataset
```

Delete useless columns , .drop is used once with inplace will remove the unwanted columns

```python
#corona_dataset_csv.drop(["Lat","Long"],axis=1,inplace=True) #will return a dataframe, set axis to 1 to specify that we want to drop the column names specified
corona_dataset_csv.head(10)
```

Aggregating the rows by country

```python
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum()
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape #returns the number of countries and the dates of our time period
```

Visualizing data related to a country for example China

```python
corona_dataset_aggregated.loc["China"].plot() #return a pandas series 
corona_dataset_aggregated.loc["Italy"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
plt.legend()
```

Calculate a good measure which represents a number which can describe the spread of the virus

```python
corona_dataset_aggregated.loc["China"][:3].plot()
```

Calculating and plotting the first derivative of the curve which shows the change in infection rate day in our period

```python
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
corona_dataset_aggregated.loc["China"].diff().max() #returns max infection cases
```

Finding maximum infection rates for all of the countries

```python
countries = list(corona_dataset_aggregated.index) #get indices of the data frame
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["max_infection_rate"] = max_infection_rates
corona_dataset_aggregated.head()
```

Create a new dataframe with only needed column

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rate"])
corona_data.head()
```
